regresión de los precios de las casas con las siguientes variantes:
 * a. Mínimos cuadrados con expansión polinomial de diferentes grados.
 * b. Mínimos cuadrados con expansión polinomial de grado 20 y penalización por norma ℓ1 y ℓ2
   con diferentes valores de $\lambda$.
 * c. Mínimos cuadrados con expansión polinomial de grado 2 y selección de atributos.
 
Grafica el error cuadrático medio en entrenamiento y validación con respecto al grado del polinomio,
valor de $\lambda$ y número de atributos. Todos los modelos deberán ser evaluados con 10 repeticiones de
validación cruzada de 5 particiones.

```python
from sklearn.datasets import load_boston
boston = load_boston()
x = boston.data
y = boston.target
```

```python
def error(y, y_hat):
    return np.square(y-y_hat).mean()
```

```python
#Definimos la fucion de validacion de 5 particiones con 10 repeticiones
#particiones = 5
#repeticion = 10
def validacion(x, y,repeticion=10,particiones=4):
    for i in range(repeticion):
        permutacion = np.random.permutation(x.shape[0])
        x = x[permutacion,:]
        y = y[permutacion]
        particion = int(x.shape[0] / particiones)
        for k in range(particiones):
            validacion = np.zeros(x.shape[0], dtype=bool)
            validacion[particion*k:particion*(k+1)] = True
            entrenamiento = np.logical_not(validacion)
            x_train = x[entrenamiento, :]
            y_train = y[entrenamiento]
            x_valid = x[validacion, :]
            y_valid = y[validacion]
            yield i, k, i * particiones + k, x_train, y_train, x_valid, y_valid
```

```python
class EscaladoEstandar:
    def fit(self, X):
        self.media = X.mean(axis=0).reshape(1,-1)
        self.std = X.std(axis=0).reshape(1,-1)
    
    def transform(self, X):
        return (X - self.media) / self.std

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)
```

```python
#definimos nuestra funcion de expansion polinomial
def polinomio(x, grado):
    f = x
    for i in range(grado - 1):
        potencia = np.power(x, i + 2)
        f = np.column_stack((f, potencia))
    
    unos = np.ones(f.shape[0])
    return np.column_stack((unos, f))
```

```python
def entrenamiento(modelo, x_train, y_train, x_valid, y_va, grado):
    es = EscaladoEstandar()
    x_train = polinomio(es.fit_transform(x_train), grado)
    x_valid = polinomio(es.fit_transform(x_valid), grado)

    modelo.fit(x_train, y_train)

    y_hat_train = modelo.predict(x_train)
    y_hat_valid = modelo.predict(x_valid)

    perdida_train = error(y_train,y_hat_train)
    perdida_valid = error(y_valid,y_hat_valid)    
    return perdida_train, perdida_valid
```

```python
class RegresionLineal():
    def fit(self, x, y):
        self._estima_parametros(x, y)
  
    def predict(self, x):
        return x @ self.parametros
```

```python
class MinimosCuadradosQR(RegresionLineal):
    def _estima_parametros(self, x, y):
        q, r = np.linalg.qr(x)
        self.parametros = np.linalg.inv(r) @ q.T @ y
```

```python
grados = np.arange(1,7,1)
ecm_ent_grados = np.zeros((grados.shape[0], 40))
ecm_val_grados = np.zeros((grados.shape[0], 40))
for i,g in enumerate(grados):
    for k,particiones,l, x_train, y_train, x_valid, y_valid in validacion(x, y,repeticion=10,particiones=4):
        rl = MinimosCuadradosQR()
        ecm_ent_grados[i,l], ecm_val_grados[i,l] = entrenamiento(rl, x_train, y_train,x_valid, y_valid, g)
        
        ecm_ent_grados[i,l], ecm_val_grados[i,l] = entrenamiento(rl, x_train, y_train,
                                                              x_valid, y_valid, g)
        print(ecm_ent_grados[i,l], ecm_val_grados[i,l])
```

```python
ent_media_rl = np.mean(ecm_ent_grados, axis=1)
ent_std_rl = np.std(ecm_ent_grados, axis=1)

val_media_rl = np.mean(ecm_val_grados, axis=1)
val_std_rl = np.std(ecm_val_grados, axis=1)
    
plt.fill_between(grados, ent_media_rl - ent_std_rl,
                  ent_media_rl + ent_std_rl, alpha=0.1,
                  color="r")
plt.fill_between(grados, val_media_rl - val_std_rl,
                  val_media_rl + val_std_rl, alpha=0.1,
                  color="b")

plt.plot(grados, ecm_ent_grados.mean(axis=1), 'o-', color="r",
         label="Entrenamiento")
plt.plot(grados, ecm_val_grados.mean(axis=1), 'o-', color="b",
         label=u"Validación")
plt.xticks(grados)
plt.xlabel('Grados de polinomio (complejidad)')
plt.ylabel('Error cuadrático medio')
plt.ylim([0, 2000])
plt.legend()
plt.show()
```
