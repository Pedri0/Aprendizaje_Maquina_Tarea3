# Prediccion de precios de casas

A partir de la base de datos de precios de casas de Boston1 (Boston Housing Data), realiza la
regresión de los precios de las casas con las siguientes variantes:
 * a. Mínimos cuadrados con expansión polinomial de diferentes grados.
 * b. Mínimos cuadrados con expansión polinomial de grado 20 y penalización por norma ℓ1 y ℓ2
   con diferentes valores de $\lambda$.
 * c. Mínimos cuadrados con expansión polinomial de grado 2 y selección de atributos.
 
Grafica el error cuadrático medio en entrenamiento y validación con respecto al grado del polinomio,
valor de $\lambda$ y número de atributos. Todos los modelos deberán ser evaluados con 10 repeticiones de
validación cruzada de 5 particiones.

```python
#importamos librerias útiles
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
#cargamos los datos
casas = pd.read_csv('housing.data',header=None,sep='\s+')
#mostramos la base de datos
casas
```

De acuerdo a la descripcion de la base de datos tenemos que cada columna representa:
Attribute Information:

   1. CRIM      per capita crime rate by town
   2. ZN        proportion of residential land zoned for lots over 
                 25,000 sq.ft.
   3. INDUS     proportion of non-retail business acres per town
   4. CHAS      Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
   5. NOX       nitric oxides concentration (parts per 10 million)
   6. RM        average number of rooms per dwelling
   7. AGE       proportion of owner-occupied units built prior to 1940
   8. DIS       weighted distances to five Boston employment centres
   9. RAD       index of accessibility to radial highways
   10. TAX      full-value property-tax rate per $\$$ 10,000
   11. PTRATIO  pupil-teacher ratio by town
   12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
   13. LSTAT    % lower status of the population
   14. MEDV     Median value of owner-occupied homes in $\$$ 1000's

```python
#renombramos las columnas que recien creamos
casas.columns=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
#mostramos la base de datos
casas
```

```python
#convertimos nuestras variables independientes a un array numpy
x = (casas.drop(columns=['MEDV'])).to_numpy()
#convertimos nuestra variable dependiente a un array numpy
y = (casas['MEDV']).to_numpy()
```

```python
#definimos la funcion de error
def error(y, prediccion):
    return np.square(prediccion-y).mean()
```

```python
#definimos la funcion de validacion de 5 particiones con 10 repeticiones, que come las variables
#x, y, repeticiones y particiones
def validacion(x, y, repeticiones=10,particiones=5):
    #ciclo para realizar las repeticiones
    for i in range(repeticiones):
        #creamos un array de numeros aleatorios que serviran como indices
        permutacion = np.random.permutation(x.shape[0])
        # definimos x como todas las filas cuyo indice contiene el array, y tomamos para cada una de ellas los valores
        #de todas las columnas
        x = x[permutacion,:]
        #tomamos los valores y de los x que ya se habian tomado
        y = y[permutacion]
        #definimos el valor de la particion (el cacho del pastel), este cacho debe ser un entero
        particion = int(x.shape[0] / particiones)
        #recorremos cada cacho o particion
        for paso in range(particiones):
            #creamos una matriz de ceros que contendrá que valores de x e y que seran para la validacion, por esa razon
            #es booleana
            validacion = np.zeros(x.shape[0], dtype=bool)
            #los valores que para validacion tendran un True en esta matriz, estos valores dependen del tamaño de la 
            #particion y del lugar de la particion en la que nos encontramos
            validacion[particion*paso:particion*(paso+1)] = True
            #Para definir los valores de entrenamiento simplemente tomamos "el inverso" de la validacion
            #ya que los valores de validacion y entrenamiento son disjuntos
            train = np.logical_not(validacion)
            #definimos nuestras x e y para validacion y entrenamiento segun las matrices anteriores
            x_train = x[train, :]
            y_train = y[train]
            x_valid = x[validacion, :]
            y_valid = y[validacion]
            #regresamos los valores de interes
            yield i, paso, i * particiones + paso, x_train, y_train, x_valid, y_valid
```

```python
#creamos una clase "escalado" para hacer un escalado de estandarizacion de los datos de entrada de la funcion, x no debe
#ser un escalar porque la funcion explota
class escalado:
    #calcula valores necesarios para calcular el escalado (media y desviacion estandar)
    def fit(self, x):
        self.media = x.mean(axis=0).reshape(1,-1)
        self.std = x.std(axis=0).reshape(1,-1)
    #calcula el escalado usando los valores de la media y desviacion
    def transform(self, x):
        return (x - self.media) / self.std
    #regresa el escalado de los valores de x
    def fit_transform(self, x):
        self.fit(x)
        return self.transform(x)
```

```python
#definimos la funcion de expansion polinomial de los valores de entrada de x para un grado definido por la entrada grado
def polinomio(x, grado):
    #phi es una variable auxiliar que contendrá el valor de la expansion polinomial
    phi = x
    #iteramos para poder realizar la expansion con el grado deseado
    for i in range(grado - 1):
        #elevamos la entrada x a la potencia i+2 
        potencia = np.power(x, i + 2)
        #apilamos en phi en forma de columna las potencias calculadas no se como explicar esto formal pero intuitivamente
        #se apila el vector de entrada y despues se transpone, como apilar o pegar vectores columna
        phi = np.column_stack((phi, potencia))
    #creamos un array de unos con la longitud de las filas de phi
    unos = np.ones(phi.shape[0])
    #regresamos el pegado del array de unos con phi, esto se hace para poder realizar la estimacion de los parametros
    #de una forma vectorial B_hat = R^-1@Q^T@y donde B es el vector columna de parametros R y Q es la descomposivion
    #QR de x y y es el vector columna de salidas
    return np.column_stack((unos, phi))
```

```python
#definimos nuestra funcion de entrenamiento y validacion cruzada que come los parametros modelo que es el modelo
#que es regresion lineal con o sin normas l1 y l2, x_train datos x de entrenamiento, y_train datos y de entrenamiento
#x_valid e y_valid son los datos x e y de validacion respectivamente, grado el grado al que se desea hacer expansion
#polinomial. Los datos seran proporcionados por la funcion validacion pero esto se hace mas adelante
def entrenamiento(modelo, x_train, y_train, x_valid, y_valid, grado):
    #instanciamos nuestra clase escalado como es
    es = escalado()
    #redefinimos nuestros datos de entrenamiento como la expansion polinomial de grado grado de los datos
    #escalados
    x_train = polinomio(es.fit_transform(x_train), grado)
    x_valid = polinomio(es.transform(x_valid), grado)
    #usamos la variable modelo para hacer un fit de nuestros datos de entrenamiento
    modelo.fit(x_train, y_train)
    #definimos las variables de prediccion de nuestro modelo para los datos de entrenamiento y validacion
    prediccion_train = modelo.predict(x_train)
    prediccion_valid = modelo.predict(x_valid)
    #calculamos los errores de la prediccion y los reales y los guardamos en variables error_train o error_valir
    #segun sea el caso, esto usando la funcion de error que definimos previamente
    error_train = error(y_train, prediccion_train)
    error_valid = error(y_valid, prediccion_valid) 
    #regresamos los valores de error calculados
    return error_train, error_valid
```

```python
#Creamos una clase de regresion lineal
class RegresionLineal():
    #estima los parametros B usando una clase hija definida aqui abajo
    def fit(self, x, y):
        self._estima_parametros(x, y)
    #realiza y regresa el producto matricial de los datos de entrada x con los parametros calculados
    def predict(self, x):
        return x @ self.parametros
```

```python
#define clase hija de minimos cuadrados sin normas
class MinimosCuadradosQR(RegresionLineal):
    #define funcion estima parametros para estimar parametros B
    def _estima_parametros(self, x, y):
        #se hace la descompasicion QR de la matriz de entrada x
        q, r = np.linalg.qr(x)
        #estima los parametros segun la definicion B_hat = R^-1@Q^T@y
        self.parametros = np.linalg.inv(r) @ q.T @ y
```

```python
#crea un array de 1 a 26 con paso 1, esto para hacer expansion polinomial de grado 1 hasta grado 26
grados = np.arange(1,26,1)
#crea matrices que almacenaran los errores de cada particion para cada repeticion, la matriz sera de 26 filas
#ya que cada fila contendra la potencia i y tendra 50 columnas, ya que son en total 10x5 valores de error de la
#validacion y entrenamiento
matrizError_train_grad = np.zeros((grados.shape[0], 50))
matrizError_valid_grad = np.zeros((grados.shape[0], 50))
#itera sobre los grados
for i,g in enumerate(grados):
    # las variables antes del in tomaran los valores retornados por la funcion validacion para un grado dado
    #estos valores dependeran de los ciclos for dentro de la funcion validacion
    for paso,particiones,l, x_train, y_train, x_valid, y_valid in validacion(x, y, repeticiones=10, particiones=5):
        #definimos nuestro modelo como minimos cuadrados con descomposicion QR
        rl = MinimosCuadradosQR()
        #los valores de los errores calculados tomaran las posiciones [i,l] segun el orden de return de la
        #funcion entrenamiento
        matrizError_train_grad[i,l], matrizError_valid_grad[i,l] = entrenamiento(rl, x_train, y_train,x_valid, y_valid, g)
```

```python
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_train_grad
#y guardamos estos valores en variables train_media y train_desviacion
train_media = np.mean(matrizError_train_grad, axis=1)
train_desviacion = np.std(matrizError_train_grad, axis=1)
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_valid_grad
#y guardamos estos valores en variables valid_media y valid_desviacion
valid_media = np.mean(matrizError_valid_grad, axis=1)
valid_desviacion = np.std(matrizError_valid_grad, axis=1)
    

#para la grafica instruimos que pinte las mantas de error para ambas matrices usando los valores de
#la media y desviavion que acabamos de calcular
plt.fill_between(grados, train_media - train_desviacion,train_media + train_desviacion, alpha=0.3,color="r")
plt.fill_between(grados, valid_media - valid_desviacion,valid_media + valid_desviacion, alpha=0.3,color="b")
#graficamos
plt.plot(grados, matrizError_train_grad.mean(axis=1), '*-', color="r",
         label="Entrenamiento")
plt.plot(grados, matrizError_valid_grad.mean(axis=1), '*-', color="b",
         label=u"Validación")
plt.xticks(grados)
plt.xlabel('Grados de polinomio')
plt.ylabel('Error cuadrático medio')
plt.ylim([0, 10000])
plt.legend()
plt.show()

plt.fill_between(grados, train_media - train_desviacion,train_media + train_desviacion, alpha=0.3,color="r")
plt.fill_between(grados, valid_media - valid_desviacion,valid_media + valid_desviacion, alpha=0.3,color="b")

plt.plot(grados, matrizError_train_grad.mean(axis=1), '*-', color="r",
         label="Entrenamiento")
plt.plot(grados, matrizError_valid_grad.mean(axis=1), '*-', color="b",
         label=u"Validación")
plt.xticks(grados)
plt.xlabel('Grados de polinomio')
plt.ylabel('Error cuadrático medio')
plt.ylim([0, 2000])
plt.xlim([1, 7])
plt.legend()
plt.show()
```

Veamos que para la expansion polinomial de grado 5 o superior el error medio cuadratico se dispara exponencialmente para el caso de la validacion mientras que a partir de grado 9 el error de entrenamiento tambien se dispara.
Buscamos que ambos sean considerablemente bajos y no sean muy diferentes entre si, por lo que una expansion a grado tres se considera que es un buen predictor

```python
#define clase hija de minimos cuadrados de norma L2
class MinimosCuadradosNormaL2QR(RegresionLineal):
    def __init__(self, lambda_dummy = 0.01):
        #definimos la variable lambda dummy que tomara los valores de lambda, la penalizacion
        self.lambda_dummy = lambda_dummy
    #estima parametros con norma L2
    def _estima_parametros(self, x, y):
        #creamos una matriz identidad multiplicada por la raiz de los valores de lambda
        lambdaIdentidad = np.sqrt(self.lambda_dummy * np.identity(x.shape[1]))
        #definimos una nueva variable x_L2 que sera el valor de entrada de x original concatenada con
        #la matriz identidad de lambdas
        x_L2 = np.concatenate((x, lambdaIdentidad))
        #definimos y_L2 como la concatenacion de y con un vector de ceros
        y_L2 = np.concatenate((y, np.zeros(x.shape[1])))
        #descomposicion qr de x_L2
        q, r = np.linalg.qr(x_L2)
        #calculamos parametros
        self.parametros = np.linalg.inv(r) @ q.T @ y_L2
```

```python
#creamos un array de potencias de 10 que comienza de 1e-10 y termina en 1e10 de 25 puntos
lambdas = np.geomspace(1e-10, 1e10, num=25)
#crea matrices que almacenaran los errores de cada particion para cada repeticion, la matriz sera de 25 filas
#ya que cada fila contendra el valor de lambda_i y tendra 50 columnas, ya que son en total 50 valores de error de la
#validacion y entrenamiento
matrizError_train_lambdas = np.zeros((lambdas.shape[0],50))
matrizError_valid_lambdas = np.zeros((lambdas.shape[0],50))
#iteramos sobre el array de lambdas recien creado
for i,la in enumerate(lambdas):
    # las variables antes del in tomaran los valores retornados por la funcion validacion para un grado dado
    #estos valores dependeran de los ciclos for dentro de la funcion validacion
    for paso,particiones,l, x_ent, y_ent, x_val, y_val in validacion(x, y, repeticiones=10, particiones=5):
        #definimos nuestro modelo de regresion lineal con norma L2 a la funcion le damos de comer las lambdas
        #contenidas en el array de lambdas
        regularizacionL2 = MinimosCuadradosNormaL2QR(lambda_dummy=la)
        #los valores de los errores calculados tomaran las posiciones [i,j] segun el orden de return de la
        #funcion entrenamiento, aqui definimos el grado como 20
        matrizError_train_lambdas[i,l], matrizError_valid_lambdas[i,l] = entrenamiento(regularizacionL2,x_ent, y_ent,x_val, y_val, 20)
```

```python
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_train_grad
#y guardamos estos valores en variables train_media_l2 y train_desviacion_l2
train_media_l2 = np.mean(matrizError_train_lambdas, axis=1)
train_desviacion_l2 = np.std(matrizError_train_lambdas, axis=1)
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_valid_grad
#y guardamos estos valores en variables valid_media_l2 y valid_desviacion_l2
valid_media_l2 = np.mean(matrizError_valid_lambdas, axis=1)
valid_desviacion_l2 = np.std(matrizError_valid_lambdas, axis=1)

#para la grafica instruimos que pinte las mantas de error para ambas matrices usando los valores de
#la media y desviavion que acabamos de calcular
plt.fill_between(lambdas, train_media_l2 - train_desviacion_l2,train_media_l2 + train_desviacion_l2, alpha=0.1,color="r")
plt.fill_between(lambdas, valid_media_l2 - valid_desviacion_l2,valid_media_l2 + valid_desviacion_l2, alpha=0.1,color="b")
#graficamos
plt.title('Regresion multivariada con penalizacion L2 ycon expansion polinomial de grado 20, con un valor maximo del eje y de 2E20 \n')
plt.plot(lambdas, train_media_l2, 'o-', color="r",
         label="Entrenamiento")
plt.plot(lambdas, valid_desviacion_l2, 'o-', color="b",
         label=u"Validación")
plt.xscale('log')
plt.xlabel('Factor de penalización $\lambda$')
plt.ylabel('Error cuadrático medio')
plt.legend()
plt.ylim([0,2e20])
plt.show()

plt.title('Regresion multivariada con penalizacion L2 ycon expansion polinomial de grado 20, con un valor maximo del eje y de 2E14 \n')
plt.plot(lambdas, train_media_l2, 'o-', color="r",
         label="Entrenamiento")
plt.plot(lambdas, valid_desviacion_l2, 'o-', color="b",
         label=u"Validación")
plt.xscale('log')
plt.xlabel('Factor de penalización $\lambda$')
plt.ylabel('Error cuadrático medio')
plt.legend()
plt.ylim([0,2000])
plt.show()

plt.title('Regresion multivariada con penalizacion L2 ycon expansion polinomial de grado 20, con un valor maximo del eje y de 10 \n')
plt.plot(lambdas, train_media_l2, 'o-', color="r",
         label="Entrenamiento")
plt.plot(lambdas, valid_desviacion_l2, 'o-', color="b",
         label=u"Validación")
plt.xscale('log')
plt.xlabel('Factor de penalización $\lambda$')
plt.ylabel('Error cuadrático medio')
plt.legend()
plt.ylim([4,10])
plt.show()
```

Veamos que para el caso de norma L2 y un polinomio de grado 20 el error de validacion es extremadamente grande aproximadamente de 1e20 y el error de entrenamiento se mantiene bajo con un valor maximo de 500 por lo que este modelo nos es inutil sin importar que factor de panalizacion usemos. Pero que pasa si reducimos el grado de la expansion polinomial a 5?

```python
#creamos un array de potencias de 10 que comienza de 1e-10 y termina en 1e10 de 25 puntos
lambdas = np.geomspace(1e-10, 1e10, num=25)
#crea matrices que almacenaran los errores de cada particion para cada repeticion, la matriz sera de 25 filas
#ya que cada fila contendra el valor de lambda_i y tendra 50 columnas, ya que son en total 50 valores de error de la
#validacion y entrenamiento
matrizError_train_lambdas = np.zeros((lambdas.shape[0],50))
matrizError_valid_lambdas = np.zeros((lambdas.shape[0],50))
#iteramos sobre el array de lambdas recien creado
for i,la in enumerate(lambdas):
    # las variables antes del in tomaran los valores retornados por la funcion validacion para un grado dado
    #estos valores dependeran de los ciclos for dentro de la funcion validacion
    for paso,particiones,l, x_ent, y_ent, x_val, y_val in validacion(x, y, repeticiones=10, particiones=5):
        #definimos nuestro modelo de regresion lineal con norma L2 a la funcion le damos de comer las lambdas
        #contenidas en el array de lambdas
        regularizacionL2 = MinimosCuadradosNormaL2QR(lambda_dummy=la)
        #los valores de los errores calculados tomaran las posiciones [i,j] segun el orden de return de la
        #funcion entrenamiento, aqui definimos el grado como 5
        matrizError_train_lambdas[i,l], matrizError_valid_lambdas[i,l] = entrenamiento(regularizacionL2,x_ent, y_ent,x_val, y_val, 5)
```

```python
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_train_grad
#y guardamos estos valores en variables train_media_l2 y train_desviacion_l2
train_media_l2 = np.mean(matrizError_train_lambdas, axis=1)
train_desviacion_l2 = np.std(matrizError_train_lambdas, axis=1)
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_valid_grad
#y guardamos estos valores en variables valid_media_l2 y valid_desviacion_l2
valid_media_l2 = np.mean(matrizError_valid_lambdas, axis=1)
valid_desviacion_l2 = np.std(matrizError_valid_lambdas, axis=1)
#para la grafica instruimos que pinte las mantas de error para ambas matrices usando los valores de
#la media y desviavion que acabamos de calcular  
plt.fill_between(lambdas, train_media_l2 - train_desviacion_l2,train_media_l2 + train_desviacion_l2, alpha=0.1,color="r")
plt.fill_between(lambdas, valid_media_l2 - valid_desviacion_l2,valid_media_l2 + valid_desviacion_l2, alpha=0.1,color="b")
#graficamos
plt.title('Regresion multivariada con penalizacion L2 ycon expansion polinomial de grado 5, con un valor maximo del eje y de 1000 \n')
plt.plot(lambdas, train_media_l2, 'o-', color="r",
         label="Entrenamiento")
plt.plot(lambdas, train_desviacion_l2, 'o-', color="b",
         label=u"Validación")
plt.xscale('log')
plt.xlabel('Factor de penalización $\lambda$')
plt.ylabel('Error cuadrático medio')
plt.legend()
plt.ylim([0,1000])
plt.show()
```

Para expansion con grado 5 se observa que con un factor de penalizacion menor a 10 el error tanto de validacion como de entrenamiento a parentan ser muy similares y se encuentran por debajo de 100. Sin embargo la manta de incertidumbre de la validacion es grande superando el error de 1000. A partir de 10e4 el error de entrenamiento crece casi linealmente con respecto a el factor de penalizacion, mientras que el error de validacion se mantiene. Con esto vemos que incrementar el grado de la expansion afecta significativamente el error de validacion.

```python
#define clase hija de minimos cuadrados de norma L1
class MinimosCuadradosNormaL1QR(RegresionLineal):
    #definimos la variable lambda dummy que tomara los valores de lambda, la penalizacion
    def __init__(self, lambda_dummy = 0.01):
        self.lambda_dummy = lambda_dummy
    #estima parametros con norma L2
    def _estima_parametros(self, x, y):
        #creamos una matriz identidad multiplicada por los valores de lambda
        lambdaIdentidad = (self.lambda_dummy * np.identity(x.shape[1]))
        #definimos una nueva variable x_L1 que sera el valor de entrada de x original concatenada con
        #la matriz identidad de lambdas
        x_L1 = np.concatenate((x, lambdaIdentidad))
        #definimos y_L1 como la concatenacion de y con un vector de ceros
        y_L1 = np.concatenate((y, np.zeros(x.shape[1])))
        #descomposicion qr de x_L1
        q, r = np.linalg.qr(x_L1)
        #calculamos parametros
        self.parametros = np.linalg.inv(r) @ q.T @ y_L1
```

```python
#creamos un array de potencias de 10 que comienza de 1e-10 y termina en 1e10 de 25 puntos
lambdas = np.geomspace(1e-10, 1e10, num=25)
#crea matrices que almacenaran los errores de cada particion para cada repeticion, la matriz sera de 25 filas
#ya que cada fila contendra el valor de lambda_i y tendra 50 columnas, ya que son en total 50 valores de error de la
#validacion y entrenamiento
matrizError_train_lambdas = np.zeros((lambdas.shape[0],50))
matrizError_valid_lambdas = np.zeros((lambdas.shape[0],50))
#iteramos sobre el array de lambdas recien creado
for i,la in enumerate(lambdas):
    # las variables antes del in tomaran los valores retornados por la funcion validacion para un grado dado
    #estos valores dependeran de los ciclos for dentro de la funcion validacion
    for paso,particiones,l, x_ent, y_ent, x_val, y_val in validacion(x, y, repeticiones=10, particiones=5):
        #definimos nuestro modelo de regresion lineal con norma L1 a la funcion le damos de comer las lambdas
        #contenidas en el array de lambdas
        regularizacionL1 = MinimosCuadradosNormaL1QR(lambda_dummy=la)
        #los valores de los errores calculados tomaran las posiciones [i,j] segun el orden de return de la
        #funcion entrenamiento, aqui definimos el grado como 20
        matrizError_train_lambdas[i,l], matrizError_valid_lambdas[i,l] = entrenamiento(regularizacionL2,x_ent, y_ent,x_val, y_val, 20)
```

```python
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_train_grad
#y guardamos estos valores en variables train_media_l2 y train_desviacion_l2
train_media_l2 = np.mean(matrizError_train_lambdas, axis=1)
train_desviacion_l2 = np.std(matrizError_train_lambdas, axis=1)
#calculamos la media y la desviacion estandar poara cada columna l de la matriz MatrizError_valid_grad
#y guardamos estos valores en variables valid_media_l2 y valid_desviacion_l2
valid_media_l2 = np.mean(matrizError_valid_lambdas, axis=1)
valid_desviacion_l2 = np.std(matrizError_valid_lambdas, axis=1)

#para la grafica instruimos que pinte las mantas de error para ambas matrices usando los valores de
#la media y desviavion que acabamos de calcular
plt.fill_between(lambdas, train_media_l2 - train_desviacion_l2,train_media_l2 + train_desviacion_l2, alpha=0.1,color="r")
plt.fill_between(lambdas, valid_media_l2 - valid_desviacion_l2,valid_media_l2 + valid_desviacion_l2, alpha=0.1,color="b")
#graficamos
plt.title('Regresion multivariada con penalizacion L1 ycon expansion polinomial de grado 20, con un valor maximo del eje y de 2E20 \n')
plt.plot(lambdas, train_media_l2, 'o-', color="r",
         label="Entrenamiento")
plt.plot(lambdas, valid_desviacion_l2, 'o-', color="b",
         label=u"Validación")
plt.xscale('log')
plt.xlabel('Factor de penalización $\lambda$')
plt.ylabel('Error cuadrático medio')
plt.legend()
plt.ylim([0,2e20])
plt.show()

plt.title('Regresion multivariada con penalizacion L1 ycon expansion polinomial de grado 20, con un valor maximo del eje y de 2E18 \n')
plt.plot(lambdas, train_media_l2, 'o-', color="r",
         label="Entrenamiento")
plt.plot(lambdas, valid_desviacion_l2, 'o-', color="b",
         label=u"Validación")
plt.xscale('log')
plt.xlabel('Factor de penalización $\lambda$')
plt.ylabel('Error cuadrático medio')
plt.legend()
plt.ylim([0,2e18])
plt.show()

plt.title('Regresion multivariada con penalizacion L1 ycon expansion polinomial de grado 20, con un valor maximo del eje y de 1000 \n')
plt.plot(lambdas, train_media_l2, 'o-', color="r",
         label="Entrenamiento")
plt.plot(lambdas, valid_desviacion_l2, 'o-', color="b",
         label=u"Validación")
plt.xscale('log')
plt.xlabel('Factor de penalización $\lambda$')
plt.ylabel('Error cuadrático medio')
plt.legend()
plt.ylim([4,1000])
plt.show()
```

Con la norma L1 vemos que sucede algo analogo al caso de la norma L2 con grado 20. Sin embargo gracias a esta norma el error de validacion disminuye drasticamente pero no lo suficiente como para considerar este modelo, ya que aun se mantiene en aproximadamente 1e18
